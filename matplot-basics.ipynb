# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from random import sample
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

### functional method

```python
# basic plot
plt.plot(x,y, 'r-1')

# label and title
plt.title('This is title')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

### multi plot

```python
#subplot arguments (num row, num column, number of plot referring)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g')


plt.subplot(1,2,2)
plt.plot(y,x, 'y')
plt.show()
```

# object oriented method

 the main idea is create figure objects

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1,0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('this is Xlabel')
axes.set_ylabel('This is Ylabel')
axes.set_title('This is set Title')
plt.show()
```

```python
fig = plt.figure()

# fig.add_axes([ [left,bot,right(width),top(height)] ])
axes1 = fig.add_axes([0.1,0.1,0.9,0.8])
axes2 = fig.add_axes([0.2,0.4,0.3,0.3])

axes1.plot(x,y)
axes1.set_title('large plot')
axes2.plot(y,x)
axes2.set_title('small plot')
plt.show()
```

```python
# fig,axes = plt.subplots(nrows=3,ncols=3)
# plt.tight_layout()
# axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

# index plot
axes[0].plot(x,y)
axes[0].set_title('First Index Title')

axes[1].plot(y,x)
axes[1].set_title('Index 1 Title')

plt.tight_layout()
```

```python
axes
```

# Figure Size and DPI

```python
#  dpi means dots per inch or pixel per inch
fig = plt.figure(figsize=(3,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.set_title('Title')
ax.set_xlabel('X Labels')
ax.set_ylabel('Y Labels')
ax.plot(x,y)
plt.show()
```

```python
#  legends - with legends we can use it to clarify the plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X sqrt')
ax.plot(x,x**3, label = 'X Cubed')

ax.legend()
plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.show()
```

#  Plot formatting

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="g", linewidth=3, alpha=0.5,)# you can use rgb hex code too for custom color
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="purple", linewidth=3,linestyle='--')# there are many other options here and you can check it in linestyles matplotlib documentation
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="y", lw=3, marker="o")# marker you can use it when you have many data points and there are many more options too
plt.show()
```

# Plot range

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="y", lw= 3 , ls='--')

# lets show the plot between 0 and 1 in x axis xlim([lowerlimit and upperlimit])
ax.set_xlim([0,1])
ax.set_ylim([0,2])#([upperlimit,lowerlimit])
plt.show()
```

# special plot types

with special plot types we can create, such as scatter plots, barplots,histogram and much more

```python
plt.scatter(x,y)
plt.show()
```

```python
data = sample(range(1,1000),100)
plt.hist(data)
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
# rectagular box plot
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```
